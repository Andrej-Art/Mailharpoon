# Data Preprocessing

Before building models, we need to prepare the data

```python
# some imports

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import pickle
```

```python
dataset = "/Users/andrejartuschenko/Desktop/mailharpoon/backend/data/UCI_training_dataset.csv"
df = pd.read_csv(dataset)
df.columns = df.columns.str.lower() #lowercase the column names
```

```python
# label encoding actually -1 = phishing and 1 = legitimate
# i want to encode thme to: 1 = phishing and 0 = legitimate better for Recall, F1 and ROC-AUC


df = df.copy()
df["result_bin"] = (df["result"] == -1).astype(int)  # 1 phishing, 0 legit
```

```python
# full feature set 

X_full = df.drop(columns=["result", "result_bin"])
y = df["result_bin"]
```

```python
# url-only feature set

url_features = [
    "having_ip_address",
    "url_length",
    "shortining_service",
    "having_at_symbol",
    "double_slash_redirecting",
    "prefix_suffix",
    "having_sub_domain",
    "https_token",
]

X_url = df[url_features]
```

```python
print("X_full columns contain result_bin?", "result_bin" in X_full.columns)
print("X_url columns contain result_bin?", "result_bin" in X_url.columns)
```

```python
#split of data set

X_train_full, X_test_full, y_train, y_test = train_test_split(
    X_full, y, test_size=0.2, stratify=y, random_state=42
)

# URL subset split via same indices
X_train_url = X_url.loc[X_train_full.index]
X_test_url  = X_url.loc[X_test_full.index]
```

```python
#overview about the split distribution and the number of features 

print(X_train_full.shape, X_test_full.shape)
print(X_train_url.shape, X_test_url.shape)
print(y_train.value_counts(normalize=True))
print(y_test.value_counts(normalize=True))
```

```python
# save preprocessed data as pkl


with open("/Users/andrejartuschenko/Desktop/mailharpoon/backend/data/processed/preprocessed_data.pkl", "wb") as f:
    pickle.dump({
        "X_train_full": X_train_full,
        "X_test_full": X_test_full,
        "X_train_url": X_train_url,
        "X_test_url": X_test_url,
        "y_train": y_train,
        "y_test": y_test
    }, f)
```
